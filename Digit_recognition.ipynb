```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# There are 1797 images with 8x8 dimensions
print("Data Shape", digits.data.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Visualising top-5 images from dataset
plt.figure(figsize=(12,12))

for index, (image,label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title("Image Data : %i \n" %label, fontsize=15)
```

```python
# Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=0)
```

```python
# Loading Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
# Creating Instance of Logistic Regression
reg = LogisticRegression()
```

```python
# Train the dataset
reg.fit(X_train, y_train)
```

```python
# Making prediction on Test Dataset
predictions = reg.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)

score = reg.score(X_test, y_test)
```

```python
# Plotting Confusion Matrix on Actual & Predicted Data
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(12,12))
sns.heatmap(cm, annot=True, linewidths=.5, square=True, cmap="Pastel1")
plt.ylabel("Actual Label")
plt.xlabel("Predicted Label")
title = "Accuracy Score : {0} ".format(score)
plt.title(title, size=12)
```
